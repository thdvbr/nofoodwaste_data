#  Introduction

In retail terms, shrinkage refers to a company's percent loss resulting from damage, product expiration and theft of unsold products. In this notebook, we will look into a dataset from a Metronom GmbH, a wholesale and food specialist company. How big is the shrinkage? Is shrinkage worse in some product categories than in others? How frequent does shrinkage occur in different products? On which days does shrinkage occur most commonly? Having insights on these various conditions helps the company to prevent the loss and increase the profitability. 

Our aims were to answer these questions by exploring the shrinkage data in different angles. Furthermore, we will try to find any existing trends or patterns with sales and shrinkage. We will plot this outcome for clear visual presentation. 

#  Data, materials and methods

We will be working on a dataset provided by Metronom, which is from one of their retailer(depot) in Germany, containing shrinkage, stocks, sales data of their product catalogue in January 2020. The dataset is in german. 

```python
import numpy as np 
import pandas as pd

from urllib import request

import matplotlib.pyplot as plt
import seaborn as sns 

from dateutil.parser import parse
from datetime import date
import calendar
```

## 1. Loading the data 
    
Let's load our data in using pandas: 

```python
filepath = "./data/Code data_METRONOM.csv"
df = pd.read_csv(filepath, encoding = "ISO-8859-1", error_bad_lines=False, sep="\t")
```

## 2. Data Wrangling
### 1) Renaming the columns

```python
df.info()
```

The column names above are not easy to understand. First, we will rename these column names into more recognizable labels according to the information Metronom provided in prior to cleaning. 

```python
df = df.rename(columns={"CATMAN_BUY_DOMAIN_ID": "purchasing_area_id",
                        "CATMAN_BUY_DOMAIN_DESC": "purchasing_area_description",
                        "PCG_MAIN_CAT_ID": "main_product_group_id",
                        "PCG_SUB_CAT_ID": "product_subgroup_category_id",
                        "PCG_SUB_CAT_DESC": "product_subgroup_category_description",
                        "PCG_CAT_ID": "product_group_category_id",
                        "PCG_CAT_DESC": "product_group_category_description",
                        "Bv NNEK" : "shrinkage_in_euros",
                        "Bv Stk" : "shrinkage_in_pieces",
                        "WA Stück" : "sales_in_pieces",
                        "Umsatz VK Netto" : "sales_in_euros",
                        "Bestand Ist NNEK" : "stock_in_euros",
                        "Bestand Ist Stück" : "stock_per_day"})
```

```python
df.columns.tolist()
```

### 2) Tidying up fields in the data


Let's look into a sample set of the df and its datatypes.

```python
df.dropna(subset=["shrinkage_in_euros"], inplace=False)
```

```python
df.dtypes
```

#### Incorrect formatting in `shrinkage_in_euros`
From the table above, we can notice that in `shrinkage_in_euros` column, commas are being used, whereas in different columns periods are being used. We will replace these commas in `shrinkage_in_euros` to periods. Also the datatype of `shrinkage_in_euros` are currently object. We will change the datatype to numerical data so we can use them properly in the analysis.

```python
df["shrinkage_in_euros"].replace(',','.',inplace = True, regex = True)
```

```python
df[["shrinkage_in_euros"]] = df[["shrinkage_in_euros"]].apply(pd.to_numeric)
```

### 3) Adding new features
#### New feature ① : `weekday` 

`DAY_DESC` column includes day data with dates but the formatting makes it difficult to use for analysis. We figured out that, by using `date_of_day` we can extract the data on which day shrinkage or sales occur and this will be useful for our analysis. We will add a new feature called `weekday`.  

```python
df["date_of_day"] = pd.to_datetime(df["date_of_day"], format = "%d.%m.%y")
df["weekday_number"] = df["date_of_day"].dt.dayofweek
df["weekday"] = df["weekday_number"].apply(lambda x: calendar.day_name[x])
```

#### New feature ② : `freshness`

Metronom stated their products can be categorized into two big categories, 'fresh' and 'ultrafresh'. There are 5 This is called `freshness` criteria. This information is not included in this dataset so we will add this categorical column on our dataframe. 

```python
df = df.assign(freshness = 'fresh')
df.loc[df['purchasing_area_description'] == 'ACM FLEISCH', 'freshness'] = 'ultrafresh'
df.loc[df['purchasing_area_description'] == 'ACM FRISCHFISCH', 'freshness'] = 'ultrafresh'
df.loc[df['purchasing_area_description'] == 'ACM GEMUESE', 'freshness'] = 'ultrafresh'
df.loc[df['purchasing_area_description'] == 'ACM BLUMEN', 'freshness'] = 'ultrafresh'
df.loc[df['purchasing_area_description'] == 'ACM OBST', 'freshness'] = 'ultrafresh'

df
```

#### New feature ③ : `shrinkage`

I will add another categorical column `shrinkage` and give it a value 1 if it has shrinkage and 0 if it does not have.

```python
df = df.assign(shrinkage = 0)
df.loc[df['shrinkage_in_euros'] < 0, 'shrinkage'] = 1
```

### 4) Dropping unnecessary columns

For convenience we will drop the columns that we do not need for the analysis. For categorical columns, we will only use `purchasing_area_description`, `product_subgroup_category_description`, `ART_NO` and `ART_NAME`. 

For the numerical columns, we won't be using any stock data as the way it has been recorded on the dataset differs from shrinkage and sales. Each row means the stock recorded by the end of the day, thus we cannot use it for comparing the sum of sales, shrinkage and stock. We decided to use `shrinkage_in_euros` and `sales_in_euros` instead of pieces, not only because of missing/non matching values but also we need a unifying unit that can be used for comparing the whole product lines.

```python
df = df.drop(['store_id','store_desc','DAY_DESC','purchasing_area_id',
             'main_product_group_id','product_group_category_id',
              'product_subgroup_category_description','product_subgroup_category_id',
              'weekday_number','stock_in_euros','stock_per_day',
              'sales_in_pieces','shrinkage_in_pieces'], axis = 1)
df.info()
```

### 5) Dealing with missing values

NaN on datasets normally refers to missing value, and they need to be treated differently. We can observe from the table below that there are a lot of missing values on shrinkage and sales data out of 738671 rows. 

According to the experts in Metronom, NaN on these columns can be treated as 0, as it doesn't mean that the data is missing, but it was simply not recorded when sales nor shrinkage did not occur. We will convert the NaNs on `shrinkage_in_euros` into 0. 

```python
#not sure to include this code here 

gdf = df.groupby(["purchasing_area_description"])
def countna(x):
    return (x.isna()).sum()
gdf.agg({'shrinkage_in_euros' : ['count', countna,'size'], 
         'sales_in_euros' : ['count', countna, 'size']})
```

```python
df.info()
```

```python
df['shrinkage_in_euros'].fillna(0, inplace = True)
df['sales_in_euros'].fillna(0, inplace = True)
```

### 6) Converting the datatypes into right types

```python
df.dtypes
```

We will change the categorical columns that are currently in object data type to category data type for higher efficiency in sorting and data storization. 

```python
for col in ['purchasing_area_description', 'product_group_category_description',
           'ART_NO', 'ART_NAME','weekday', 'freshness', 'shrinkage']:
    df[col] = df[col].astype('category')
df.dtypes
```

### 7) Shaping the dataframe into how we want

In prior to exploratory analysis, dataframe needs to be cleaned  according to the goal. On this project we will focus on analysing the articles that had shrinkage in any of the days in January 2020. Articles that never had any shrinkage are considered as irrelvant for the project. 

```python
df["ART_NO"].value_counts() 
```

This dataframe consists of store data collected within January 2020. There were 31 days in January. We can assume that each article's data got recorded every day in January, by checking how many rows that each `ART_NO` has. From the outcome of the code above, we can see that not every article has data on all of the 31 days in january. We will only look into articles that has data on the whole month of january for analysis. 

```python
#df1 is a subset made out of items that has rows for everyday in january.
df = df.groupby("ART_NO").filter(lambda x: (x.name != 0 ) and (x.ART_NO.size == 31))
```

```python
df.shape
```

Now we have to look for articles that had shrinkage in df and store the `ART_NO` in a list called 'artno', make a new df with articles that has the same `ART_NO` in 'artno'.

```python
df1 = df[df["shrinkage_in_euros"] < 0]
artno = pd.unique(df1["ART_NO"])
artno
```

```python
new_df = []
for article in artno: 
    new_df.append(df[df.ART_NO == article])
```

```python
sdf = pd.concat(new_df)
```

```python
sdf.shape
```

### 8) Removing incorrect data




```python
sdf[sdf["shrinkage_in_euros"] > 0 ]
```

We can see there are 3 rows of shrinkage that has positive number rather than negative. According to the expert in Metronom positive shrinkage can occur from various reasons like employee mistakes, we got advised to exclude these rows from analysis.  

```python
#drop positive shrinkage 3 rows 
sdf.drop(sdf[sdf["shrinkage_in_euros"] > 0].index, inplace=True)
```

### 9) Modifying `shrinkage_in_euros` into absolute values


On this dataset, when shrinkage occur it is recorded in `shrinkage_in_euros` column as negative value. To compare these with `sales_in_euros`, we want to convert these values into their absolute values for better plotting. 

```python
sdf['shrinkage_in_euros'] = sdf['shrinkage_in_euros'].abs()
```
