```python
import numpy as np 
import pandas as pd
from urllib import request
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
filepath = "./data/Code data_METRONOM.csv"
df = pd.read_csv(filepath, encoding = "ISO-8859-1", error_bad_lines=False, sep="\t")
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
df.count()
```

```python
df = df.rename(columns={"Bv NNEK" : "Monthly Shrinkage in euros",
                   "Bv Stk" : "Monthly Shrinkage in pieces",
                   "WA Stück" : "Sales in Pieces",
                   "Umsatz VK Netto" : "Sales in euros",
                   "Bestand Ist NNEK" : "Average Stocks in Euros per Day",
                   "Bestand Ist Stück" : "Average Stocks per Day"})

df.head()
```

what are we trying to predict here ? 
the monthly shrinkage. "Monthly Shrinkage in euros" or "Monthly Shrinkage in pieces" should be our label set.  

What the columns mean: 
    
1)__store_id__ : ID of the store

2)__store_desc__: Name of the store 

3)__CATMAN_BUY_DOMAIN_ID__: Purchasing area ID 

4)__CATMAN_BUY_DOMAIN_DESC__: Purchasing area description 
ACM BACKWAREN
ACM BLUMEN
ACM FINEKOST
ACM FLEISCH
ACM FRISCHFISCH
ACM GEMUESE
ACM GOURVENIENCE
ACM KAESE
ACM MOLKEREIPRODUKTE
ACM OBST
ACM WURST

5)__PCG_MAIN_CAT_ID__: Main product group ID 

6)__PCG_CAT_ID__: Product category ID 

7)__PCG_CAT_DESC__: Product category description 

8)__PCG_SUB_CAT_ID__: Product sub category ID

9)__PCG_SUB_CAT_DESC__: Product sub category description

10)__ART_NO__: Article identifier number

11)__ART_NAME__: Article identifier name






Will drop all the rows that has missing shrinkage information.

```python
df = df.dropna(subset=["Monthly Shrinkage in euros", "Monthly Shrinkage in pieces"])
```

```python
df.count()
```

```python
df.head()
```

We won't need store_id, and store_desc as it's all the same, so will drop those columns.

We can see that for "Bv NNEK"(Monthly shrinkage in euros) they are using commas and other columns such as "Bv Stk"(Monthly shrinkage in pieces) are using periods. We will change the commas to period so we can work with them properly.

```python
df["Monthly Shrinkage in euros"].replace(',','.',inplace = True, regex = True)
```

```python
df.head()
```

```python
df.dtypes
```

Monthly Shrinkage in euros should be numerical value but it's stored as object types. We need to convert these into numeric types.

```python
df[["Monthly Shrinkage in euros"]] = df[["Monthly Shrinkage in euros"]].apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
duplicatedRows = df[df.duplicated(['ART_NAME'])]
duplicatedRows
```
